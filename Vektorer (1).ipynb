# Skalarproduktet

Hvis a og b er to vektorer i rommet, gjelder: <br>
(a) |a + b|$^{2}$ = |a|$^{2}$ + 2 a · b + |b|$^{2}$ <br>
(b) |a − b|$^{2}$ = |a|$^{2}$ − 2 a · b + |b|$^{2}$ <br>
(c) (a + b) · (a − b) = |a|$^{2}$ − |b|$^{2}$ <br>
(d) a · b = |a|·|b|· cos v, der v er vinkelen mellom vektorene a og b <br>
(e) |a + b| ≤ |a| + |b| <br>
(f) a∙b = a$_{1}$∙b$_{1}$ + a$_{2}$∙b$_{2}$ + ..... + a$_{n}$∙b$_{n}$ <br>
(g) a∙b = 0 ⇔ a ⊥ b

```python
a = array([1,2,3]) # konvertering til array
b = array([2,3,4])
print ('a =', a, 'og b =',b)
```

```python
a_dot_b = dot(a, b)
```

```python
print('Skalarproduktet: a∙b =', a_dot_b)
```

```python
vinkel_ab = vinkel(a, b)
# i python sånn som vi har satt opp med pylab import *, så regner vi alltid i radianer.
print('Vinkelen mellom vektor', a, 'og vektor', b, 'er', round(vinkel_ab,3),
      'rad, altså,', round(degrees(vinkel_ab), 2), 'grader.')
```

# Vektorproduktet

For vektorer a, b, c i rommet gjelder: <br>
(a) a × b = −(b × a) <br>
(b) a × (b + c) = a × b + a × c og (a + b) × c = a × c + b × c <br>
(c) a × (sb) = s(a × b) og (sa) × b = s(a × b) der s ∈ R <br>
(d) a × b står ortogonalt på både a og b <br>
(e) a x b = 0 ⇔ a ∥ b <br>
(f) |a x b| = |a|∙|b|∙sin(a,b) <br>
(g) |a × b|$^{2}$ = |a|$^{2}$·|b|$^{2}$ − (a · b)$^{2}$ ( Lagranges identitet) <br>
(h) a × b = [a$_{2}$∙b$_{3}$ − a$_{3}$∙b$_{2}$, a$_{3}$∙b$_{1}$ − a$_{1}$∙b$_{3}$, a$_{1}$∙b$_{2}$ − a$_{2}$∙b$_{1}$] <br>


```python
print('Vektorproduktet: axb =', cross(a, b))
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

vectors=np.array( [ [0,0,1,1,-2,0], [0,0,2,1,1,0],[0,0,3,2,1,0],[0,0,4,0.5,0.7,0]]) 
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
for vector in vectors:
    v = np.array([vector[3],vector[4],vector[5]])
    vlength=np.linalg.norm(v)
    ax.quiver(vector[0],vector[1],vector[2],vector[3],vector[4],vector[5],
            pivot='tail',length=vlength,arrow_length_ratio=0.3/vlength)
ax.set_xlim([-4,4])
ax.set_ylim([-4,4])
ax.set_zlim([0,4])
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.show()
```

```python
def draw(x, y): 
    xPlusy = (abs(x[0]+y[0]),abs(x[1]+y[1]))
    array = np.array([[0, 0, x[0], x[1]], 
                      [x[0], x[1], y[0], y[1]], 
                      [0, 0, xPlusy[0], xPlusy[1]]])
    print(array)
    X, Y, U, V = zip(*array)
    print("X =",X)
    print("Y =",Y)
    print("U =",U)
    print("V =",V)
    plt.figure()
    plt.ylabel('Y-axis')
    plt.xlabel('X-axis')
    ax = plt.gca()
    ax.quiver(X, Y, U, V, angles='xy', scale_units='xy',color=['r','b','g'],scale=1)
    ax.set_xlim([0, 10])
    ax.set_ylim([0, 10])
    plt.draw()
    plt.show()
draw([4,2],[2,4])
```
